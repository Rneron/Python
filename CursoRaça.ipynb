```python
# Importando as bibliotecas necessárias 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
dados = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/tabelas/Censo Diversidades 2022.xlsx', sheet_name = 'TUDO - CursoRaça')
dados = dados.drop(labels=179,axis=0)
dados = dados.drop(labels=178,axis=0)

```

```python
dados['Proporção AMARELA'] = dados.AMARELA / dados.TOTAL
dados['Proporção BRANCA'] = dados.BRANCA / dados.TOTAL
dados['Proporção INDÍGENA '] = dados.INDÍGENA / dados.TOTAL
dados['Proporção NÃO INFORMADA '] = dados['NÃO INFORMADA'] / dados.TOTAL
dados['Proporção PARDA '] = dados.PARDA / dados.TOTAL
dados['Proporção PRETA '] = dados.PRETA	 / dados.TOTAL
```

```python
dados = dados.replace (np.nan, 0)
```

```python
dados.to_csv('dados.csv',sep=';', encoding='iso-8859-1')
```

***Vendo os cursos com mais estudantes de cada raça, proporcionalmente e absoluto***

```python
prop_amarela = dados.nlargest(5,'Proporção AMARELA') 
top5_amarela = dados.nlargest(5,'AMARELA')
prop_amarela # proporção
```

```python
ax = sns.barplot(y=top5_amarela.AMARELA,x=top5_amarela.Curso)
ax.figure.set_size_inches(16,8)
plt.title('Cursos com o Maior Número de Pessoas Amarelas',fontdict={'fontsize': 20})
ax.set_xlabel('Curso',fontdict={'fontsize':12})
ax.set_ylabel('Total de pessoas Amarelas',fontdict={'fontsize':12})
plt.show()
```
